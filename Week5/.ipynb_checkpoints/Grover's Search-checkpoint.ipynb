# Grover's Search Algorithm

Up to this point, we've dealt with simple algorithms that serve as thought experiements for people who are new to the field of Quantum Computing to realize the true potential of this field. Now, we'll be understanding one of the most popular algorithms that's out there: Grover's Search Algorithm.

## Premise

Grover's Search is commonly demonstrated using a deck of cards or a database in the cloud somewhere. The key part of this idea is that we have some set of <u>unstructured/unsorted</u> data. Any CS student that has gone through a coding interview or a Data Structures class can tell you that it will take Linear Time to find a single element in the unsorted data or $O(n)$ runtime. This is because we have to check each element one-by-one to verify that the element we're currently checking is the element, and if not, we move onto the next element. If the search element is not in the data, then we've exhausted the entire data set searching for the element in question.

Grover's Search functions a bit differently, and we'll use the classic 'Neddle in a Haystack' example to showcase this. So, let's say your friend tells you that they dropped a sewing needle in a haystack, and they ask you to help try to find it. The linear approach would be to check between every single straw pieces that makes up the haystack; however, that's an inefficient use of time.

The functionality of Grover's Search would be like asking your friend 'What general area did the neddle fall in?'

That way, you <u>reduce your search space</u> to the **most likely** areas that the needle could be in.

## High-level Operation

There are 3 main steps that occur in Grover's Search, one of which, we've already seen.

### The H Gate Army
First, we apply an array of H Gates across our data qubits, mathematically denoted as $H^{2\otimes}$. Like in previous algorithms, this is our way of setting all $2^N$ possibilities into equal superposition, meaning that they are equally likely to occur at the beginning.

### Inversion about the Mean - The Oracle
In our haystack example, inversion around the mean is referring to identifying areas where the needle **is more likely** to be.

When we're running our qubits through this step, we use a phase flip to change the qubit state, for example, from $\left| \psi \right\rangle$ to $-\left| \psi \right\rangle$. This step is applied to all qubits that fit the criteria of the oracle we're running our qubits through.


### Adjusting our search space - Diffusion
This step is us actually adjusting our search space to the new areas that apply. The diffusion operation has the effect of amplifying the probability of finding the target item(s) while reducing the probability of finding the other items by reducing all qubits that didn't have a phase change from step 2. 

### Repeat
For $N$ qubits used in this operation, we repeat steps 2 and 3 $\sqrt N$ times, then we measure the data qubits to find our answer.

## Mathematical Connection

In mathematics, we use the concept of orthogonality and orthonormativity to describe the relationship between the unstructured dataset and the item we're trying to find.

### Orthonormativity definition
In linear algebra, orthogonality refers to the property of two vectors being perpendicular, or "at right angles" to each other. Two vectors $u$ and $v$ are considered orthogonal if their **dot product is equal to zero**, i.e. $uÂ·v = 0$. In this case, the vectors are said to be orthonormal if they are also unit vectors, which means they have a magnitude of 1, which, in Quantum Computing, is always the case.

### Relating to the Diagram
Let's say that we have an unstructured data set where $\left| s \right\rangle$ is the superposition of all items within the set $\left| b_t \right\rangle$ is the item we're searching for.

In part B, we're conducting step 2 from the operation outline, which is called Querying the Oracle or Reflection. This is where we attach a phase shift on the qubits that fit the Oracle criteria, which
- This is denoted by $-R_t\left| s \right\rangle$

In part C, this is the diffusion operation. This operator find all qubit possibilities that didn't have a phase shift and reduces their probability from $\frac {1}{\sqrt N}$ to something much smaller, or even 0 in some cases. As a counter force, the probabilities of the phase-shifted qubits increases.
- This is denoted by $-R_sR_t\left| s \right\rangle$
- This process gets us closer to finding $\left| b_t \right\rangle$ by decreasing the angle between $\left| b_t \right\rangle$ and the closest vector

## Simple Example

```python
from qiskit import *
from qiskit.tools.visualization import plot_histogram
from qiskit.circuit.library import PhaseOracle
from qiskit.algorithms import Grover, AmplificationProblem
%matplotlib inline
```

### The Reflector Operation
For this example, we'll be using a simple C-Z gate to isolate the **winning state**, $\left| 11 \right\rangle$. This is shown as a 2-qubit system where $q_0$ is the control qubit in this context and $q_1$ is the target qubit.

```python
oracleCircuit = QuantumCircuit(2,name='oracleCircuit')
oracleCircuit.cz(0,1)
oracleCircuit.to_gate()
oracleCircuit.draw(output='mpl')
```

```python
mainCircuit = QuantumCircuit(2,2)
mainCircuit.h([0,1])
mainCircuit.append(oracleCircuit,[0,1])
mainCircuit.draw(output='mpl')
```

### The Diffusion Operator
This step is used to amplify the desired state(s). This is done by reducing the probabilities of the undesired states so that the probabilities of the desired states gets **amplified**.

```python
reflectionCircuit = QuantumCircuit(2,name="reflectionCircuit")
reflectionCircuit.h([0,1])
reflectionCircuit.z([0,1])
reflectionCircuit.cz(0,1)
reflectionCircuit.h([0,1])
reflectionCircuit.to_gate()
reflectionCircuit.draw(output='mpl')
```

### The Full Circuit
The main circuit outlines the full 3 steps as separate operations so that it's clear as to what's going on.

```python
mainCircuit.append(reflectionCircuit,[0,1])
mainCircuit.measure([0,1],[0,1])
mainCircuit.draw(output='mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
result = execute(mainCircuit,backend=backend,shots=1).result()
counts = result.get_counts(mainCircuit)
plot_histogram([counts])
```

## Qiskit Implementation Example

In this example, we're using a Qiskit class that interprets boolean statements into a quantum circuit. That quantum circuit gets run through a QASM simulator and the output qubit states describe states that make the original boolean statement true.
- The format for writing a boolean statement is as follows
    - Paranthesis for operation order
    - & is the AND operation
    - | is the OR operation
    - ~ is a negation

```python
oracle = PhaseOracle('((James & Lars) | (Kirk & Rob)) & ~(Lars & Rob)')
problem = AmplificationProblem(oracle=oracle, is_good_state=oracle.evaluate_bitstring)
backend = Aer.get_backend('qasm_simulator')
grover = Grover(quantum_instance=backend)
result = grover.amplify(problem)
```

```python
print(result.circuit_results[0])
```

```python
plot_histogram(result.circuit_results)
```

### Outcomes
- James - Lars (1100)
- James - Kirk - Lars (1011)
- Kirk - Rob (0011)
- James - Kirk - Rob (0111)

## Caveats to GS

### Not a Guarantee, it's just a Reduction
When we conduct Grover's Search Algorithm $\sqrt N$ times given a set of $N$ qubits, by the end of it, we end up with a reduced search space, but that space isn't always guaranteed to end up in a single output; it could possibly be multiple possibilities, but the important thing is that we went from $2^N$ possibilities to a smaller multitude.

### Not a One Size Fits All Solution
Grover's Search may not be ideal for all types of searches in an unstructured database. The cases where GS would not apply would be need to be researched more, but this shows that Quantum Computers will not be the dominating machine in every regard in the foreseeable future.
